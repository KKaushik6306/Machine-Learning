## Exploratory Data Analysis 

## Univariate Analysis & Bivariate Analysis 

Univariate Analysis - Where we have only one or two feature. <br>
Bivariate Analysis - Where we have 2 or 3 features. <br>
Multivariate Analysis Where we have more than 3 features.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
```

# Univariate Analysis 

```python
df_setosa = df.loc[df['species']=='setosa']
```

```python
df_virginica = df.loc[df['species']=='virginica']
df_versicolor = df.loc[df['species']=='versicolor']
```

```python
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'o')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'o')
plt.xlabel('Petal Length')
plt.show()
```

```python
sns.FacetGrid(data=df,hue='species',size=5).map(sns.distplot,'petal_length').add_legend();
plt.show()

#This Graph is specifically when you want to do analysis with one feature using hue part

#Lines are called - Probability Desnsity Function
```

```python
sns.distplot(df['sepal_length'])
```

## Bivariate Analysis 

```python
sns.FacetGrid(data=df,hue='species',size=5).map(plt.scatter,'petal_length','sepal_width').add_legend();
plt.show()

#This Graph is specifically when you want to do bivariate analysis between two numerical feature.
```

```python
sns.jointplot(x='petal_length',y='sepal_width',data=df,kind='hex')
```

# Multivariate Analysis 

```python
sns.pairplot(data=df,hue='species')
```
